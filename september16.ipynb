```python
import yfinance as yf

spy_data = yf.download('SPY', start='2022-01-01')
spy_data.head()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
plt.title('SPY Price Over Time')
plt.plot(spy_data['Close'], label='SPY Close Price')
plt.show()
```

```python
spy_data['SMA50'] = spy_data['Close'].rolling(window=50).mean()

plt.figure(figsize=(12, 6))
plt.title('SPY Price with 50-Day SMA')
plt.plot(spy_data['Close'], label='SPY Close Price')
plt.plot(spy_data['SMA50'], label='50-Day SMA', color='orange')
plt.legend()
plt.show()
```
